```python
from googleapiclient.discovery import build
from pytube import YouTube
import os

# ---------- YOUR API KEY ----------
API_KEY = "AIzaSyC0DsPu74T7eoeyb3jnteiFPkr7ohonHxY"  

# ---------- SEARCH FUNCTION USING API ----------
def search_youtube(query, max_results=5):
    youtube = build("youtube", "v3", developerKey=API_KEY)

    request = youtube.search().list(
        q=query,
        part="snippet",
        type="video",
        maxResults=max_results
    )
    response = request.execute()

    print("\nðŸ” Search Results:")
    videos = []
    for i, item in enumerate(response['items'], start=1):
        title = item['snippet']['title']
        video_id = item['id']['videoId']
        print(f"{i}. {title}  --> https://youtube.com/watch?v={video_id}")
        videos.append(video_id)

    return videos

# ---------- DOWNLOAD FUNCTION USING pytube ----------
def download_video(video_url, path="downloads"):
    if not os.path.exists(path):
        os.makedirs(path)

    print("\nâ¬‡ Downloading Video...")
    yt = YouTube(video_url)
    stream = yt.streams.get_highest_resolution()
    stream.download(output_path=path)
    print(f"âœ… Download Completed: {stream.title}\n")
    return stream.title

# ---------- MAIN PROGRAM ----------
if __name__ == "__main__":
    print("ðŸŽ¬ YouTube Video Downloader with API\n")

    search_query = input("Enter topic to search: ")

    # Step 1 â†’ Search YouTube
    video_ids = search_youtube(search_query)

    # Step 2 â†’ Select a video
    choice = int(input("\nSelect video number to download: "))
    selected_video_id = video_ids[choice - 1]

    video_url = f"https://www.youtube.com/watch?v={selected_video_id}"

    # Step 3 â†’ Download video
    download_video(video_url)
```
